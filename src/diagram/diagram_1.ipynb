```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import sklearn as sk

# charger les données
df=pd.read_csv('../data/new_dataset.csv')

def get_survived_color(level_hobisity):
    if level_hobisity == "Normal_Weight":
        return "red"
    if level_hobisity == "Overweight_Level_I":
        return "orange"
    if level_hobisity == "Overweight_Level_II":
        return "blue"
    if level_hobisity == "Obesity_Type_I":
        return "green"
    if level_hobisity == "Insufficient_Weight":
        return "black"
    if level_hobisity == "Obesity_Type_II":
        return "yellow"
    if level_hobisity == "Obesity_Type_III":
        return "gray"

for row in df.itertuples():
    plt.scatter(row.Height, row.Weight, color=get_survived_color(row.NObeyesdad))


# Définir une fonction linéaire (par exemple, y = 2x + 3)
def fonction_lineaire(x):
    return 30*(x**2)

# Générer des valeurs pour l'axe x
x_valeurs = np.linspace(1.3, 2.1, 100)

# Calculer les valeurs correspondantes sur l'axe y en utilisant la fonction linéaire
y_valeurs = fonction_lineaire(x_valeurs)

# Création du graphe
# plt.legend(labels=['Normal_Weight', 'Overweight_Level_I', 'Overweight_Level_II', 'Obesity_Type_I', 'Insufficient_Weight', 'Obesity_Type_II', 'Obesity_Type_III'])
plt.plot(x_valeurs, y_valeurs, label='Droite linéaire')

```
