```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#### Titre ####

df = pd.read_csv('../data/dataset.csv')

# Valeurs de nos classes cibles
classes = ['Obesity_Type_III', 'Normal_Weight', 'Overweight_Level_I', 'Insufficient_Weight', 'Obesity_Type_I', 'Overweight_Level_II', 'Obesity_Type_II']

# Visualisation pour montrer la répartition des classes NObeyesdad en nombre
plt.figure(figsize=(17, 6))
plt.subplot(1, 2, 1)

NObeyesdad_counts = df["NObeyesdad"].value_counts()
sns.barplot(x=NObeyesdad_counts.index, y=NObeyesdad_counts.values, palette='Set2', order=classes)
plt.title("Distribution of NObeyesdad Classes", fontweight="black", size=14, pad=15)

for i, v in enumerate(NObeyesdad_counts.values):
    plt.text(i, v, v, ha="center", fontsize=14)

plt.xticks(rotation=90)

# Visualisation pour montrer la répartition des classes NObeyesdad en pourcentage
plt.subplot(1, 2, 2)
colors = sns.color_palette('Set2', len(classes))
plt.pie(NObeyesdad_counts, labels=classes, autopct="%.2f%%", textprops={"size": 14},
        colors=colors, startangle=90)
center_circle = plt.Circle((0, 0), 0.3, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.title("Distribution of NObeyesdad Classes", fontweight="black", size=14, pad=15)
plt.show()
```
